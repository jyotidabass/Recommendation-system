## Latent Factor Models

An approach that tries to explain the ratings by characterizing both users and items by uncovering **latent features** 

**Question**

What are latent features?

Set of factors, very specific to the domain, but not necessarily obvious or measured explicitly.

One class of models that does this is matrix factorization

## Non-negative Matrix Factorization

- Characterize both items and users by vectors of factors inferred from item-rating patterns  
- Matrix type of input data
- One dimension representing user
- The other representing items
- A metric to measure loss (typically squared error)

### Evolution of Matrix Factorization

![](img/mfevolve.png)

#### Basic NMF Model

- Map both users and items to a joint latent factor space
- user-item interactions are inner products in that space

**Decomposition of the rating matrix**  
R -> m users, n movies  
Row in U -> user's affinity to the features (m users, f latent factors/features)  
Row in M -> item's relation to the features (n movies, f latent factors/features)  

![](img/rmu.png)


- U indicates how much user likes f latent factors  
- M indicates how much one item obtains f latent factors
- The dot product indicates how much user likes item  
- The decomposition automatically ranks features by their impact on the ratings
- features may not be intuitive though !
- Model has hyperparameters (regularization, learning rate)


**Common Learning Algorithms**

- stochastic gradient descent
- alternating least squares
